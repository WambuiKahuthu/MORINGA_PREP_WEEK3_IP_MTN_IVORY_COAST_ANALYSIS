##Selecting Data To Be Used

###Merging the three days datasets

```
#merge the three telecom datasets by columns
Telcomsetall = pd.concat([Telcomset1, Telcomset2, Telcomset3], axis = 0)

#display the dataframe
Telcomsetall
```

###Dropping unnecessary columns




```
#drop unwanted columns from the merged telecom dataset
Telcomsetall.drop(['INDEX','CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B'], axis=1, inplace =True)
Telcomsetall
```

```
#drop unnecessary columns from cells geo dataframe
geoset.drop(['INDEX', 'STATUS', 'LOCALIZATION', 'DECOUPZONE', 'ZONE_NAME', 'LONGITUDE', 'LATITUDE','AREA','CELL_ID'], axis=1, inplace=True)
#display the dataframe
geoset
```

##Cleaning The Datasets

###Dropping duplicates

```
#drop duplicates in the merged telecom dataframe
Telcomsetnodups = Telcomsetall.drop_duplicates()
#display dataframe with no duplicates
Telcomsetnodups
```

```
#drop duplicates in cells geo dataframe
geosetnodups = geoset.drop_duplicates()
#display dataframe with no duplicates
geosetnodups
```

###Dropping rows with missing values

```
#drop rows with missing values from the merged telecom dataset with no duplicates
Telcomset = Telcomsetnodups.dropna()
Telcomset
```

```
#drop rows with missing values from cells geo dataset with no duplicates
cellsgeo = geosetnodups.dropna()
cellsgeo
```

##Merging The Datasets

```
#merge the merged telcom dataset and the cells geo data set by the common column
TelecomGeo = pd.merge(Telcomset, cellsgeo, on='SITE_ID', how='inner')

#change 'DATE_TIME' column's data type to datetime  
TelecomGeo["DATE_TIME"] = pd.to_datetime(TelecomGeo["DATE_TIME"])

#display dataframe
TelecomGeo
```

##Formatting The Dataset

```
#sort the dataset by ascending order of "DATE_TIME"
TelecomGeo = TelecomGeo.sort_values(by="DATE_TIME", ascending=1)
#display the sorted dataframe
TelecomGeo
```

###order of regions by their values

```
#group "VALUE" and "REGION" then convert back to dataframe
valueregion = TelecomGeo['VALUE'].groupby(TelecomGeo['REGION']).sum().reset_index()
#sort dataframe in descending order
valueregion.sort_values('VALUE', ascending =0)
```

###order of cities by their values

```
#group "VALUE" and "CITY" then convert to dataframe
valuecity = TelecomGeo['VALUE'].groupby(TelecomGeo['CITY']).sum().reset_index()
#sort dataframe in descending order of "VALUE"
valuecity.sort_values('VALUE', ascending =0)
```

###regions in which the value gained is above average

```
#find the mean of "VALUE" in the valueregion dataframe
valueregion['VALUE'].mean()
#find regions that have values above average
topregions = valueregion[valueregion['VALUE'] > valueregion['VALUE'].mean()]
#sort dataframe by descending order of "VALUE"
TopRegions = topregions.sort_values(by='VALUE', ascending=0)
#display sorted dataframe
TopRegions
```
